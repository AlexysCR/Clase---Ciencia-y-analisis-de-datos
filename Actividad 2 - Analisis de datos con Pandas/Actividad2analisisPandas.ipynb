8.	Ejecuta el siguiente código y comenta qué hace cada línea:

```python
# Separa la la columna de "ReadingDateTime", separa la información por los espacios en blanco y guarda la infromación en un nuevo dataframe "datetime_df"
datetime_df = air_df.ReadingDateTime.str.split(' ',expand=True)
# Renombra las columnas del nuevo data frame: "Date" y "Time"
datetime_df.columns = ['Date','Time']
datetime_df

# Toma la columna "Date" de "datetime_df", separa la infromación por "/" y lo guarda en un nuevo data frame "date_df"
date_df = datetime_df.Date.str.split('/',expand=True)
# Renombra las columnas del nuevo data frame: "Day", "Month" y "Year"
date_df.columns = ['Day','Month','Year']
date_df

# Une los data frames "date_df", "datetime_df" y "air_df", y a su vez elimina las columnas "ReadingDateTime" y "Year" y guarda el resultado en el data frame previo "air_df" 
air_df = air_df.join(date_df).join(datetime_df.Time).drop(columns=['ReadingDateTime','Year'])
# Renombra las columnas del nuevo data frame: "Month", "Day", "Time" y "Species"
air_df = air_df.set_index(['Month','Day','Time','Species'])
air_df
```

10.	Explica cuáles son las diferencias y similitudes entre el par `melt()/pivot()` y el par `stack()/unstack()`. Si tuvieras que elegir una contraparte para `melt()` entre `stack()/unstack()`, ¿cuál elegirías? ¿Por qué?

**Ambos métodos "melt() y pyvot()" y "stack()/unstack() reestructuran data frames, sin embargo actuan de maneras distintas:**

**melt()/pivot() son métodos contrarios uno de otro, así como stack()/unstack()**

* **melt(): Convierte un data frame de formato ancho a largo (apila columnas en filas)**
* **pyvot(): Convierte un data frame de formato largo a ancho (apila filas en columnas)**
* **stack(): Convierte columnas en un índice jerarquico (apila columnas en un índice)**
* **unstack():Convierte un índice jerarquico en columnas (desapila un índice en columnas)**

**En caso de tener que elegir entre los diferentes métodos, para éste caso, yo elegiría stack() ya que condesna la información de mejor manera y se evita repetición de la información aunque los registros horizontales incrementan. Pero a cambio obtenemos mejor legibilidad y un output horizontal de menor tamaño.**
